# Early Fusion

```python
earlyfusion = NYdf.append([Sdf, Ldf])
earlyfusion
```

# Clean Data

```python
import string
import nltk
import re
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
def remove_pattern(input_txt):
    input_txt = re.sub('[^a-zA-Z]',' ',input_txt)
    input_txt = input_txt.lower()
    input_txt = input_txt.split()
    input_txt = [stemmer.stem(word)for word in input_txt]
    input_txt = [t for t in input_txt if len(t)>2]
    input_txt = ' '.join(input_txt)
    return input_txt

earlyfusion[['educationInfoForAgeGroupEstimation', 'workInfoForAgeGroupEstimation']] = earlyfusion[['educationInfoForAgeGroupEstimation', 'workInfoForAgeGroupEstimation']].fillna('')
for i in ['educationInfoForAgeGroupEstimation', 'workInfoForAgeGroupEstimation']:
    earlyfusion[i] = earlyfusion[i].apply(lambda x: remove_pattern(x))

earlyfusion
```

```python
earlyfusion['gender'] = earlyfusion['gender'].fillna('')
#earlyfusion['relaionship'] = earlyfusion['relationship'].fillna('', inplace=True)
#earlyfusion['educationLevel'] = earlyfusion['educationLevel'].fillna('')
earlyfusion['income'] = earlyfusion['income'].fillna('')

earlyfusion['relaionship'] = earlyfusion['relationship'].fillna(earlyfusion.relationship.dropna().max(),inplace=True)
earlyfusion['educationLevel'] = earlyfusion['educationLevel'].fillna(earlyfusion.educationLevel.dropna().max())

earlyfusion
```

```python
earlyfusion_info = earlyfusion.educationInfoForAgeGroupEstimation + earlyfusion.workInfoForAgeGroupEstimation
xtrain, xtest, ytrain, ytest = train_test_split(earlyfusion_info
                                                , earlyfusion[['gender', 'relationship', 'educationLevel', 'income']]
                                               , test_size = 0.3)
```

# Vectorize Data

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(max_df = 0.90, min_df = 2, stop_words = 'english', max_features = 1000)
xtrain_vec = vect.fit_transform(xtrain)
xtest_vec = vect.transform(xtest)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vec = TfidfVectorizer(max_df = 0.90, min_df = 2, stop_words ='english', max_features = 1000)
xtrain_tfid = tfidf_vec.fit_transform(xtrain)
xtest_tfid = tfidf_vec.transform(xtest)
```

# Prediction

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,f1_score
from sklearn.multiclass import OneVsRestClassifier


LogReg_pipeline = Pipeline([('clf', OneVsRestClassifier(LogisticRegression(solver='sag'), n_jobs=-1)),])
for i in ['gender', 'relationship', 'educationLevel', 'income']:
    print('**Processing {} comments...**'.format(i))
    LogReg_pipeline.fit(xtrain_vec, ytrain[i])
    prediction = LogReg_pipeline.predict(xtest_vec)
    print('Test accuracy is {}'.format(accuracy_score(ytest[i], prediction)))
    print('Test F1_score is {}'.format(f1_score))
    print("\n")
```
